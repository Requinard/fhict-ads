```python
import pandas as pd
import numpy as np
import matplotlib

%matplotlib inline
```

```python
df = pd.read_csv('data.csv')
print(df)
```

```python
df['date_fixed'] = [pd.to_datetime(d) for d in df['date']]
```

```python
df.groupby('l_ipn').plot.line(x='date_fixed', y='f', title='Connections per day per workstation')
```

```python
df.groupby('l_ipn').plot.line(x='date_fixed', y='f', title='Connections per day per workstation, seen logarithmically', logy=True)
```

```python
df.plot(kind='scatter', x='r_asn', y='f', figsize=(50,15))
```

```python
df_remote_to_connections = df.pivot_table(index='r_asn', values='f', aggfunc=len).sort_values(by='f', ascending=False)
df_remote_to_connections[:100].plot(kind='bar', title='Remote ip\'s that received the most connections', figsize=(50, 15))
```

```python
ips_to_remote = df.pivot_table(index='l_ipn', values='r_asn', aggfunc=len).sort_values(by='r_asn', ascending=False)

ips_to_remote.plot(kind='bar', title='Amount of remote ips a workstation connected to')
```
