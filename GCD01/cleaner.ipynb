```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv('data.csv')
df
```

# Cleaning  up the gender column

We check if there are any matches to male or female. We drop the other items

```python
df['What is your gender?'].apply(lambda x: "male" if any(map(lambda w: w in x.lower(), ("man", "male"))) else x).apply(lambda y: "female" if any(map(lambda w: w in y.lower(), ("vrouw", "female"))) else y).apply(lambda x: None if x != 'male' and x !='female' else x)
```

# Cleaning up the trouser column

We try to match a regex

```python
number_filter = re.compile('(\d+)')

def find_num(x):
    matcher = number_filter.match(x)
    if matcher:
        return str(matcher.group(0))
    else:
        return None
df['How many pairs of trousers do you own?'].astype(str).apply(find_num)
```
